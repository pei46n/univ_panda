```python
import os
import pandas as pd
```

```python
path = '../ds_business_design.csv'
df = pd.read_csv(path, header=None)
df.head()
```

```python
points = df[8].values
```

```python
a = points[0]
b = a.split(', ')
b
```

```python
new = []

for point in points:
    new.append(point.split(', '))
```

```python
p_df = pd.DataFrame(new, columns=[8, 9])
p_df.head()
```

```python
dropdf = df.drop([8], axis=1)
dropdf.head()
```

```python
f_df = pd.merge(dropdf, p_df, left_index=True, right_index=True)
# f_df
# f_df.to_csv('../ds_business_design_.csv', header=None, index=None)
```

```python
f_df.head()
```

```python
import folium



names = f_df[1].values
lats = f_df[8].values
lngs = f_df[9].values



map = folium.Map(location=[36, 140], zoom_start=5)


for name, lat, lng in zip(names, lats, lngs):
    folium.Marker(
        location=[lat, lng],
        popup=name,
        icon=folium.Icon(color='blue')
    ).add_to(map)

map.save('./panda/static/all_stadiums.html')
map
```
